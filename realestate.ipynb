```python
# 평균값 구하기
temp = []
for i in dfAvgPrice.loc[:, '12월':'9월.1'].values:
    temp.append(i.mean())
    
result = pd.DataFrame(temp, index=dfAvgPrice['시군구'], columns=['평균가'])
result.loc['제주']
```

```python
temp = {}

dfAvgArea['분양가격(㎡)'] = dfAvgArea['분양가격(㎡)'].str.replace(',', '')
dfAvgArea['분양가격(㎡)'] = pd.to_numeric(dfAvgArea['분양가격(㎡)'], errors='coerce')
dfAvgArea = dfAvgArea.dropna(subset=['분양가격(㎡)'])
dfAvgArea['분양가격(㎡)'] = dfAvgArea['분양가격(㎡)'].astype(int)
```

```python
dfAvgArea[['지역명', '분양가격(㎡)']].groupby('지역명').mean().loc['제주']
```

```python
temp = dfAvgArea[['지역명', '분양가격(㎡)']].groupby('지역명').mean().loc['서울']
temp1 = dfAvgArea[['지역명', '분양가격(㎡)']].groupby('지역명').mean().loc['경기'] + dfAvgArea[['지역명', '분양가격(㎡)']].groupby('지역명').mean().loc['인천']

print('서울 평균 분양가격 : {}\n수도권 평균 분양가격 : {}'.format(int(temp.values[0]), int(temp1.values[0])))
```
