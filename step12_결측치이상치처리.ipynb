**결측치 확인**

1. isnull()과 sum()을 연결한 메소드 체인 사용
2. isnull() : 결측치가 있을 경우 True <-> notnull()
3. df.isnull().sum() : 결측치의 개수
4. df.isnull().mean() : 결측치의 비율
5. dropna() : 결측치 제거
- 결측치(NaN)를 포함하는 레코드(행) 제거
- how=‘any’ : 결측치가 하나라도 있으면 제거 (기본 옵션)
- how=‘all’ : 모든 행이 결측치인 경우 제거
- dropna() 사용시 index를 따로 설정하지 않은 경우 reset_index(drop=True)를
메서드 체인으로 연결해서 사용하는 것을 추천

**결측치 대체에 가장 많이 사용되는 방법**
- 평균 또는 중앙값 또는 최빈값을 사용
- 결측치를 꼭 대체해야하는 것일까? 평균과 중앙값, 최빈값 중 어느 것을 사용할까?
- Trial and error method
- 결측치 대체에 사용가능한 다른 방법은 없을까?
- 빈번한 값으로 대체
- 분포에 따른 이상치 파악
- 범범주형 변수의 결측치 대체 : 최빈값으로 대체/ ‘UNK’ 등 새로운 범주의 생성
- KNN 즉 비슷한 유형으로 대체
- 범주형인 경우 많이 나오는 것으로 대체하기도 함
