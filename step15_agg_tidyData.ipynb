```python
import numpy as np
import pandas as pd
```

```python
flights = pd.read_csv('input/flights.csv')
flights
```

```python
flights.columns
```

```python
flights.groupby('AIRLINE').agg({'ARR_DELAY' : 'mean'}).head()
```

```python
flights.groupby('AIRLINE')['ARR_DELAY'].agg('mean').head()
```

```python
flights.groupby(['AIRLINE', 'WEEKDAY'])
```

```python
flights.WEEKDAY
```

```python
group_cols = ['ORG_AIR', 'DEST_AIR']
agg_dict = {'CANCELLED':['sum', 'mean'] , 'AIR_TIME':['mean', 'var']}
air_info = flights.groupby(group_cols, as_index=False).agg(agg_dict)
air_info
```

```python
air_info.columns
```

```python
level0 = air_info.columns.get_level_values(0)
level0
```

```python
level1 = air_info.columns.get_level_values(1)
level1
```

```python
air_info.columns = level0 + '_' + level1
air_info
```

```python
fruit = pd.read_csv('input/fruit.csv', index_col='State')
fruit
```
